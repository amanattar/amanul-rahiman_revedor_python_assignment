# Data Processing 

```python
#importing the libraries

import pandas as pd
import json
```

```python
#importing the dataset

df = pd.read_json('data.json', lines=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#rename the columns
df.rename(columns={'a': 'address', 'liid': 'linkedin_id', 'linkedin':'profile_url', 'n':'name', 't':'mob_number', 'e':'email'}, inplace=True)
```

```python
df.head()
```

```python
# remove the values with null address

df = df[pd.notnull(df['address'])]
```

```python
df = df.reset_index()
```

```python
df.isnull().sum()
```

```python
df.shape
```

# Creating Unique ID 

```python
data_sample = df
```

```python
data_sample['id'] = ""
```

```python
for i in range(len(data_sample['linkedin_id'])):
    data_sample['id'][i] = i+1
```

```python
data_sample
```

```python
#creating data_main

data_main = data_sample[['id','linkedin_id','address','profile_url']]
```

```python
data_main.isnull().sum()
```

```python
#creating data_email

data_email = data_sample[['id','linkedin_id','email']]
```

```python
data_email
```

```python
# Deleteing Null email rows
data_email = data_email[pd.notnull(data_email['email'])]
```

```python
data_email.isnull().sum()
```

```python
data_email
```

```python
#creating data_number

data_number = data_sample[['id','linkedin_id','mob_number']]
```

```python
# Deleting Null mobile number
data_number = data_number[pd.notnull(data_number['mob_number'])]
```

```python
data_number
```

```python
# creating csv files

data_main.to_csv("data_main.csv",index=False)
data_email.to_csv("data_email.csv",index=False)
data_number.to_csv("data_number.csv",index=False)
```
